# Exploratory Data Analysis (EDA)
TechX Sales Dataset

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
pd.set_option('display.max_columns',None)
```

```
df=pd.read_csv('TechX_Sales_Dataset (1).csv')
df.head()
```

```
df['Order_Date']=pd.to_datetime(df['Order_Date'])
```

```
plt.figure(figsize=(8,5));sns.histplot(df['Sales'],bins=30,kde=True);plt.show()
```

```
plt.figure(figsize=(8,5));sns.boxplot(x=df['Sales']);plt.show()
```

```
plt.figure(figsize=(8,5));sns.scatterplot(data=df,x='Sales',y='Profit');plt.show()
```

```
plt.figure(figsize=(8,6));sns.heatmap(df.select_dtypes(include=np.number).corr(),annot=True,cmap='coolwarm');plt.show()
```

```
sns.pairplot(df.select_dtypes(include=np.number));plt.show()
```

```
plt.figure(figsize=(8,5));sns.barplot(data=df,x='Region',y='Sales');plt.show()
```

```
plt.figure(figsize=(8,5));sns.barplot(data=df,x='Category',y='Sales');plt.show()
```

```
df['Month']=df['Order_Date'].dt.month
df.groupby('Month')['Sales'].sum().plot(marker='o');plt.show()
```

```
df.groupby('Product')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(10,5));plt.show()
```
